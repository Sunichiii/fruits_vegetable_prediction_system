Importing Libraries

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

Test Set Image Processing

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    r'D:\Python\Projects\Fruits and vegetable prediction system\Edible Dataset\test',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

Loading Model

```python
cnn = tf.keras.models.load_model('trained_mode_version1.h5')
```

```python
# Loading the model
cnn = tf.keras.models.load_model('trained_mode_version1.h5')

# Recompiling the model 
cnn.compile(
    optimizer='adam', 
    loss='categorical_crossentropy', 
    metrics=['accuracy']  
)
```

Visualizing and performing single prediction on single image

```python
#Test Image Visualization
import cv2
image_path = r'D:\Python\Projects\Fruits and vegetable prediction system\Edible Dataset\test\raddish\Image_2.jpg'
# Reading an image in default mode
img = cv2.imread(image_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #Converting BGR to RGB
# Displaying the image 
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

Testing Model

```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Converts single image to a batch.
predictions = cnn.predict(input_arr)
```

```python
print(predictions)
```

```python
result_index = np.argmax(predictions) #Return index of max element 
print(result_index)
```
